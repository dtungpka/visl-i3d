```python

rgb = pickle.load(open('test_18096_raw.pkl', 'rb')) #[logit, classif,predicted, labels]
depth = pickle.load(open('test_depth_raw.pkl', 'rb'))

result = []

for i in range(len(rgb)):
        data = rgb[i]
        labels = data[3]
        logit = data[0]
        logit_depth = depth[i][0]

        #combine with weights 0.5 each
        logit = (logit + logit_depth) / 2
        print(logit.shape)
        classif = np.max(logit, axis=1)
        predicted = np.argmax(classif)
        print(labels, predicted)
        result.append([logit, classif, predicted, labels])


ev.evaluate('test',1,result,class_labels=class_labels)
```
